```python
# reading the extracted csv file which contains Talcher data

ex_data = pd.read_csv('extracted_data_hour.csv')
```

```python
# displaying extracted data

ex_data # 923 rows Ã— 16 columns
```

```python
ex_data.isnull().sum() # checks how many missing data are present in each column
```

```python
# To know datatype of each column 

for col in ex_data.columns:
    print(f"{col}:", ex_data[col].dtype)
```

```python
# handling missing data
# replacing missing values using backward and forward fill
val_1 = ex_data.bfill() # backward fill
val_1.head()
```

```python
# forward fill
val_2 = val_1.ffill()
val_2.head()
```

```python
# as Xylene is neglegible so filling it with zero value

val_2['Xylene'] = val_2['Xylene'].fillna(0)
val_2.head()
```

```python
# selecting only those columns which have numericals values

num_cols = val_2.select_dtypes(include = ["float64"])
num_cols.head()
```

```python
# displaying column names of numerical values

num_cols.columns
```

```python
num = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3',
       'Benzene', 'Toluene', 'Xylene', 'AQI']
```

```python
 # checking again how many missing data are present in each column
val_2.isnull().sum()
```

```python
# Applying Tranformation and initializing StandardScaler()

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
val_2[num] = scaler.fit_transform(val_2[num])
val_2.head()
```

```python
# encoding categorical data
for col in val_2.columns:
    print(f"{col}:",val_2[col].dtype)
```

```python
cat_features = val_2.select_dtypes(include = ["object"])
cat_features.head()
```

```python
cat_features.columns
```

```python
categorical_features = ['City', 'Datetime', 'AQI_Bucket']
val_2[categorical_features].head()
```

```python
# OneHotEncoding

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore')
val_3 = pd.DataFrame(encoder.fit_transform(val_2[categorical_features]).toarray(),columns=encoder.get_feature_names_out(categorical_features))
val_3.head()
```

```python
#concatenating dataframes

transformed_data = pd.concat([val_3,val_2[num]],axis=1) #AQI_bucket comes before the numerical columns
transformed_data.head(10)
```

```python
# saving transformed data

transformed_data.to_csv('transformed_data_hour.csv',index=False)
```
